# Обучение моделей

```python
# Зададим RandomState
RANDOMSTATE = 12345
```

```python
# Функция
def split(df, features_df, target_df, test_df):
    # Разбиваем выборку в соотношении 78:22. 
    train, valid = np.split(df.sample(frac=1, random_state=RANDOMSTATE), [int(.78*len(df))])

    # Проверяем сколько строк и столбцов в полученных датасетах
    print('Размер обучающей выборки - {:.0%}'.format(train.shape[0] / data.shape[0]))
    print('Размер валидационной выборки- {:.0%}'.format(valid.shape[0] / data.shape[0]))
    
    print('Проверяем размер обучающей выборки')
    print(train.shape) 
    print('Проверяем размер валидационной выборки')
    print(valid.shape)
    
    # Разбиваем на целевой признак и признаки модели.
    target_train = train[target_df]
    features_train = train[features_df]

    target_valid = valid[target_df]
    features_valid = valid[features_df]
    
    model = LinearRegression()
    model.fit(features_train, target_train)

    predicted_target= model.predict(features_valid) #предсказание на валидационной выборке
    mean_predicted_target = predicted_target.mean()
    

    print('')
    print('Среднее значение предсказанного количества школьников за год :', predicted_target.mean())
    print('RMSE :', mean_squared_error(target_valid, predicted_target)**0.5)
    
    predicted_target= model.predict(test_df) #предсказание на валидационной выборке
    print('Предсказания на 2025-2045 годы: ')
    print(predicted_target.reshape((-1, 1)))
    
    #return target_valid, predicted_target
    
```
